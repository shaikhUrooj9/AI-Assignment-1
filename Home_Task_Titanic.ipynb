```python

df = pd.read_csv("/content/drive/MyDrive/train.csv")

```

```python
df.head()
```

```python
counts = df['Survived'].value_counts()

plt.bar(counts.index, counts.values)

plt.title("Survived vs Not Survived")
plt.xlabel("Survival Status (0 = Not Survived, 1 = Survived)")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
#● Plot the number of people that survived by both Class and Gender
sns.countplot(x='Pclass', hue='Sex', data=df[df['Survived']==1])
plt.title("Survival Count by Passenger Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Survivors")

plt.show()
```

```python
#Can we Remove Outliers or Apply Normalization on Complete Table?
#No we cannot apply normalization or remove outliers on the complete dataset because:
#The dataset contains categorical columns such as:
#Name,Sex ,Embarked
#Normalization and outlier detection are applied only to numerical columns, such as:
#Age ,Fare ,SibSp ,Parch
#Therefore preprocessing is applied only to numeric features.
```

```python
#Explore Pandas:
#Difference between Pandas Series and DataFrame (Short Answer)
#Pandas Series:
#A one-dimensional data structure that stores a single column of data with an index.
#Pandas DataFrame:
#A two-dimensional data structure that stores data in rows and columns like a table. Each column in a DataFrame is a Series.
```

```python
#● Delete or Add column in Dataframe. (Also define concept of Axis)
df['FamilySize'] = df['SibSp'] + df['Parch']
df.head()
```

```python
df=df.drop('Cabin', axis=1)
df.head()

#Axis Concept:Axis defines the direction of operation.
#Axis	Meaning axis = 0	Rows ,axis = 1	Columns
#Example: df.drop('Cabin', axis=1)
#Means remove column Cabin.
```

```python
# change column names in Dataframe:
df.rename(columns={'Pclass':'PassengerClass'}, inplace=True)
df.head()
```

```python
# Apply Logical Conditions
#Example: Display passengers whose Fare is greater than 100
df[df['Fare'] > 100]
```
